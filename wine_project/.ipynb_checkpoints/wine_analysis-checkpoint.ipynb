```python
print df_red['quality'].median()
print df_white['quality'].median()
```

```python
df_red_median = df_red.median(axis = 0)
```

```python
df_white_median = df_white.median(axis = 0)
```

```python
df_median_wine = pd.concat([df_red_median, df_white_median],axis=1)
df_median_wine.rename(columns = {0: 'Red', 1: 'White'}, inplace=True)
df_median_wine

df_median_wine.to_csv('data/median_wine.csv')
```

```python
ax = df_median_wine.plot.bar(rot=75, figsize=(50,30), fontsize=50, color=['r', 'y']);
ax.legend(fontsize=50);
ax.set_title("Red wine vs White wine Median Values", fontsize=50);
#TODO:
#fix bar graph title and legend size and color
```
